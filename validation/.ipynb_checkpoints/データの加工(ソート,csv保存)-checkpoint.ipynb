```python
import pandas as pd
import numpy as np
import datetime
```

```python
#データのソート
```

```python
df=pd.read_excel("./data/original_data_hyderabad.xlsx")
```

```python
df=df.sort_values("DateTime").reset_index().iloc[:,1:]
```

```python
len(df)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["AQI-IN"]=df["AQI IN"]
df=df.drop("AQI IN", axis=1)
```

```python
#DateTimeの秒を00にする
```

```python
df["DateTime"]=df["DateTime"].str[:-5]+"00:00"
```

```python
#欠損を補完する
```

```python
df_datetime=pd.DataFrame()
df_datetime["DateTime"]=pd.date_range(start="20191123", end="20200221", freq="H").astype("str")[:-1]
```

```python
df_datetime
```

```python
df=pd.merge(df_datetime, df, on="DateTime", how="left")
```

```python
df
```

```python
df.to_csv("data/sorted_data_hyderabad.csv")
```

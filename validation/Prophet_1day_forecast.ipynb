```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
from fbprophet import Prophet
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import requests
```

```python
url="https://api.aqi.in/getexceldata"
payload={"serialno":"PLLODA000406", "time":"4", "slottype":"2"}
r=requests.get(url, headers=payload)

jsondata=r.json()

aqi_in=jsondata["data"][1]["avrangearray"]
pm25=jsondata["data"][2]["avrangearray"]
timearray=jsondata["data"][7]["timearray"]
```

```python
#Create dataframe
df=pd.DataFrame({
    "time":timearray,
    "AQI-IN":aqi_in,
    "pm25":pm25
})


#----------Data processing----------

df["time"]=df["time"].str[:-5]+"00:00"

start_string=df.iloc[-1,0]
#datetime to string
start_datetime=datetime.datetime.strptime(start_string, '%Y-%m-%d %H:%M:%S')
#string to datetime
starttime=start_datetime.strftime("%Y%m%d")


end_string=df.iloc[0,0]
#datetime to string
end_datetime=datetime.datetime.strptime(end_string, '%Y-%m-%d %H:%M:%S')
#1 day add
end_add_datetime=end_datetime + datetime.timedelta(days=1)
#string to datetime
endtime=end_add_datetime.strftime("%Y%m%d")

#fill missing value
df_datetime=pd.DataFrame()
df_datetime["time"]=pd.date_range(start=starttime, end=endtime, freq="H").astype("str")[:-1]
df=pd.merge(df_datetime, df, on="time", how="left")

#Change type from string to datatime 
df['time'] = pd.to_datetime(df['time'])

#Set index
df = df.set_index('time')
df.head()

df["pm25"]=df["pm25"].astype("float64")
```

```python
#Length of train data(83day*24 = 1992)
#train=12/26〜3/19
#test=3/21〜3/27
len_train=1680

score=[]

#Execute accuracy evaluation seven times while shifting one day
for i in range(7):
    df_input =df.reset_index()[["time", "pm25"]].rename(columns={'time':'ds', 'pm25':'y'})
    
    #train-test split
    train=df_input.loc[0:len_train,:]
    #test=df_input.loc[len_train+1:len_train+24*7,:]
    test=df_input.loc[len_train+1:len_train+24,:]
    
    #Normalization
    train["y"]=np.log(df_input.loc[0:len_train,:]["y"])
    
    #Set cap and floor in the train data
    train["cap"]=np.log(700)
    train["floor"]=0

    #Create model
    model = Prophet(growth='logistic', daily_seasonality=False)
    model.add_seasonality(name='daily', period=7, fourier_order=90) 
    model.add_seasonality(name='weekly', period=7, fourier_order=50) 
    model.fit(train)

    #Forecast 168hours later
    future = model.make_future_dataframe(periods=24, freq="H")
    
    #Set cap and floor in the forecast data
    future["cap"]=np.log(600)
    future["floor"]=0
    
    #Forecast
    forecast = model.predict(future)
    forecast["yhat"]=np.exp(forecast["yhat"])
    
    df_forecast=pd.DataFrame(test.reset_index()["y"].values,columns=["original"])
    df_forecast["forecast"]=forecast[-168:].reset_index()["yhat"]
    df_forecast.index=test["ds"]
    
    #Plot the result(not categorical)
    plt.figure(figsize=(16,9))
    plt.ylim(0, 250) #depends on data
    plt.plot(df_forecast["original"], label="test")
    plt.plot(df_forecast["forecast"], label="forecast", color="r")
    plt.legend(fontsize=14)
    plt.show()
    
    #Change category
    df_forecast1=pd.DataFrame()
    df_forecast1["original_category"]=df_forecast["original"].apply(lambda x:aqi_numeric(x))
    df_forecast1["forecast_category"]=df_forecast["forecast"].apply(lambda x:aqi_numeric(x))
    
    #Sampling at median every 3 hours
    df_forecast1=df_forecast1.resample("3H").median()
    #Impute missing values on average
    df_forecast1["original_category"]=df_forecast1["original_category"].fillna(df_forecast1["original_category"].mean())
    
    #Plot the result
    plt.figure(figsize=(16,9))
    plt.plot(df_forecast1["original_category"], label="test")
    plt.plot(df_forecast1["forecast_category"], color="r", label="forecast")
    plt.legend(fontsize=14)
    plt.show()
    
    #Accuracy evaluation
    #"""pm2.5なのでカテゴリではなく、RMSEで評価"""
    #rmse = np.sqrt(mean_squared_error(df_forecast["original"], df_forecast["forecast"]))
    #print(rmse)
    #score.append(rmse)
    mae = mean_absolute_error(df_forecast["original"], df_forecast["forecast"])
    print(mae)
    score.append(mae)
    
    #Shifting one day
    len_train=len_train+24
    
"""print("RMSE score")
print(score)
print("Average RMSE score")
print(sum(score)/len(score))"""
print("MAE score")
print(score)
print("Average MAE score")
print(sum(score)/len(score))
```

```python
def aqi_numeric(x):
    if x>=0 and x<=50:
        return 1
    elif x>50 and x<=100:
        return 2
    elif x>100 and x<=200:
        return 3
    elif x>200 and x<=300:
        return 4
    elif x>300 and x<=400:
        return 5
    elif x>400 :
        return 6

#トレーニングデータの数(70日*24)
len_train=1680

score=[]

#1日ずらしながら7回精度評価を行う
for i in range(7):

    #train-testの分割
    ts=df.iloc[0:len_train, :]["pm25"].interpolate('linear')
    test=df.iloc[len_train:len_train+24,:]["pm25"].interpolate('linear')

    #一回差分の計算
    ts_diff=ts.diff().dropna()

    #パラメータ推定
    param=sm.tsa.arma_order_select_ic(ts_diff, ic="aic", trend="nc")
    p, q=param["aic_min_order"]

    #モデリング
    sarimax=sm.tsa.SARIMAX(ts,
                          order=(p,1,q),
                          seasonal_order=(1,1,1,24),
                           enforce_stationaly=False,
                           enforce_invertibility = False
                          ).fit()



    #現在の時間
    now=ts.index[-1]

    #１時間後から7日後まで予測するための時間の作成
    forecast_1hour=now + datetime.timedelta(hours=1)
    forecast_7day=now + datetime.timedelta(hours=24)
    #文字列に変更
    forecast_1hour=forecast_1hour.strftime("%Y-%m-%d %H:%M:%S")
    forecast_7day=forecast_7day.strftime("%Y-%m-%d %H:%M:%S")
    print("ーーーーーーーーーー"+str(forecast_1hour)+ "〜"+str(forecast_7day)+ "の予測ーーーーーーーーーー")

    #予測(引数は文字列で与える)
    sarimax_pred=sarimax.predict(forecast_1hour, forecast_7day)

    #実測値・予測値を数値カテゴリ(1~6)に変換
    df_forecast1=pd.DataFrame()

    df_forecast1["original_category"]=test.apply(lambda x:aqi_numeric(x))
    df_forecast1["forecast_category"]=sarimax_pred.apply(lambda x:aqi_numeric(x))

    # 中央値でサンプリング
    df_forecast1=df_forecast1.resample("3H").median()

    #Plot
    #Plot the result(not categorical)
    plt.figure(figsize=(16,9))
    plt.ylim(0, 250) #depends on data
    plt.plot(test, label="test")
    plt.plot(sarimax_pred, label="forecast", color="r")
    plt.legend(fontsize=14)
    plt.show()
    
    plt.figure(figsize=(16,9))
    plt.plot(df_forecast1["original_category"])
    plt.plot(df_forecast1["forecast_category"], color="r")
    plt.show()

    #精度評価
    mae = mean_absolute_error(test.values, sarimax_pred.values)
    #mae = mean_absolute_error(df_forecast1["original_category"], df_forecast1["forecast_category"])
    print(mae)
    score.append(mae)

    len_train=len_train+24

print(score)
```
